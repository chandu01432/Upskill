#Space Complexity
- is the amount of memory/space an algorithm uses as input size grows.

1. Big O - Worst Case : Most space required
2. Omega - Best Case : Least space required
3. Theta - Average Case : average space required

#Time Complexity
- is the amount of time an algorithm takes to run, based on the size of the input.
- It tells us how fast or slow your code will be when data increases.

#Types of Time Complexity:

1. O(1) - Constant Time - Always take the same time
2. O(n) - Linear Time - Time grows with size
3. O(log n) - Logarithmic Time - Time shrinks because input is divided
4. O(n^2) - Quadratic Time - Time grows fast with nested loops
5. O(nlogn) - n-Log Time - dividing and sorting and merging

```python
#Creating a Fresh Node
class Node:
  def __init__(self,data):
    self.data=data #stores value
    self.next=None #points to the next node

class LinkedList:
  def __init__(self):
    self.head=None #this keeps the track of the beginning of the list

  def traverse(self):
    current=self.head #starting from the first node
    while current: #as long as there is a current(node)
      print(current.data, end="-->") #10-->20-->
      current=current.next #updating current to the next node
    print("None")

  def IAB(self,data):
    new_node=Node(data) #creating a new node
    new_node.next=self.head #new_node points to the first node
    self.head=new_node #Head=New_node

  def IAE(self,data):
    new_node=Node(data) #creating a new node
    #Linked List empty
    if self.head is None: #If true : no list is available
       self.head=new_node #if list is empty, new_node becomes the first node
       return

    #Linked List has nodes
    current=self.head
    while current.next:
      current=current.next #traversing till the last node
    current.next=new_node #updating last node(pointing to null earlier) to new_node

  def IAP(self,data,pos):
    new_node=Node(data) #creating a new node
    #if pos=0,node to be inserted at the front
    if pos == 0:
      self.IAB(data)
      return
    #if pos != 0, something greater requires traversing
    current=self.head
    count=0 #count positions iterated
    while current and count<pos-1:
      current=current.next #TRAVERSAL
      count+=1
    if current is None: #reached end of list
      print("Position is out of range")
      return
    new_node.next=current.next
    current.next=new_node

  def DAB(self):
    if self.head is None:
      print("List is Empty")
      return
    Node_deleted=self.head #good practice but not a compulsion
    self.head=self.head.next

  def DAE(self):
    if self.head is None:
      print("List is Empty")
      return
    current=self.head
    while current.next.next:
      current=current.next
    current.next=None

  def DAP(self,pos):
    if self.head is None:
      print("List is Empty")
      return
    #Current: check for node availability
    #count: keeping a track of iterations
    #count<pos-1: traversing to the node right before pos
    current=self.head
    count=0

    if pos==0:
      self.DAB()
      return
    else:
      while current and count<pos-1:
        current=current.next
        count+=1
      if current is None:
        print("Invalid")
        return
      current.next=current.next.next

l1=LinkedList()
l1.IAB(10)
l1.IAE(20)
l1.IAP(30,1)
l1.IAB(40)
l1.DAP(1)
l1.DAB()
# l1.DAP(2)
l1.traverse()
```

#Doubly Linked Lists
1. data
2. prev(left) -> points to the previous node
3. next(right) -> points to the next node
**BIDIRECTIONAL** : goes in both forward and backward direction, makes operations faster

DLL1.png

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.prev=None
    self.next=None

class DoublyLinkedList:
  def __init__(self):
    self.head=None

  def traverse_forward(self):
    current=self.head
    while current: #while there is a node
      print(current.data, end="<-->") #print data
      current=current.next #update current
    print("None")

  def traverse_backward(self):
    current=self.head
    if current is None: #current is None: no list
      print("List is empty")
      return
    while current.next: #we're at the beginning of the list, traversing to the last node
      current=current.next  #updating current
    while current: #while there is a node
      print(current.data, end="<-->")
      current=current.prev
    print('None')

  def IAB(self,data):
    new_node=Node(data)
    if self.head is not None: #list is present
       self.head.prev=new_node #first node's prev pointing to new node
       new_node.next=self.head #new_node points to the first node
    self.head=new_node #new head value

  def IAE(self,data):
    new_node=Node(data) #creation of a new node
    if self.head is None: #checking for the list
      self.head=new_node #if no list, new_node becomes first node
      return
    current=self.head #copy of the head
    while current.next: #iteration till the last
      current=current.next #updation of current
    current.next=new_node #linking
    new_node.prev=current

  def IAP(self,data,pos):
    new_node=Node(data)
    if self.head is None: #checking for the list
      self.head=new_node #if no list, new_node becomes first node
      return

    if pos == 0: #if pos=0, Insert at beginning
      self.IAB(data) #so calling the function
      return

    current=self.head
    count=0
    while current and count<pos-1:
      current=current.next
      count+=1
    if current is None:
      print("Position is out of bounds.")
      return
    new_node.next=current.next
    new_node.prev=current
    if current.next:
      current.next.prev=new_node
    current.next=new_node

  def DAB(self):
    if self.head is None:
      print("List is empty.")
      return
    self.head=self.head.next
    if self.head:
      self.head.prev=None

  def DAE(self):
    if self.head is None:
      print("List is empty.")
      return
    if self.head.next is None: #if head.next -> None, means single present
      self.head=None #update head to None
      return
    current=self.head
    while current.next:
      current=current.next
    current.prev.next=None

  def DAP(self,pos):
    if self.head is None:
      print("List is Empty")
      return
    if pos==0:
      self.DAB()
      return
    #Current: check for node availability
    #count: keeping a track of iterations
    #count<pos-1: traversing to the node right before pos
    current=self.head
    count=0
    while current and count<pos-1:
      current=current.next
      count+=1
    if current is None:
      print("Position out of bounds")
      return

    node_to_delete=current.next
    current.next=node_to_delete.next #current.next.next

    if node_to_delete.next: #current.next.next
      node_to_delete.next.prev=current #current.next.next.prev=current
```
