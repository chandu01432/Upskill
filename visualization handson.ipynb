```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/cars (1).csv', skipinitialspace= True)
```

```python
# Subplots
# i want to create a subplot  where two subplots will be there
# 1) . mpg vs hp
# 2)  cylinders vs hp
# you can use scatterplot
```

```python
plt.subplots(1,2, figsize =(16,6))  # 1 row , 2 columns , figsize - width, height

plt.subplot(1,2,1)  # no. of rows , no.of columns , position
sns.scatterplot(data = df, x="mpg", y="hp", hue = "brand")
plt.title("Scatterplot of mpg vs hp")

plt.subplot(1,2,2)
sns.scatterplot(data = df, x="cylinders", y="hp", hue = "brand")
plt.title("Scatterplot of cylinders vs hp")

plt.show()

```

```python
sns.scatterplot(data = df, x = "mpg", y = "hp", hue = "brand")  # hue - explicitly differentiate between your values, pallete
plt.title("Scatter plot  of MPG VS HP")
plt.show()
```
