```python
rows = int(input("Enter the number of rows you want your christmas tree to be in"))
for i in range(1, rows+1):
  spaces = rows - i
  stars = 2 * i - 1
  print(" "*spaces + "*"*stars + " "*spaces)
```

```python
for i in range(1,5):
  if i == 3:
    break # This will exit the loop when i is 3
  print(i)

print("Out of loop")
```

## continue

```python
for i in range(1,5):
  if i == 3:
    continue # This will skip the loop statements when i is 3
  print(i)

print("Out of loop")
```

```python
list1 = ["24", "twenty", "67", "agje", "55"]

for i in list1:
  if not i.isdigit():
    continue
  print(i)
```

```python
n=int(input("enter a number:"))
for i in range(0,n):
  for j in range(i,n):
    print(" ",end=" ")
  for k in range(0,i):
    print("*",end=" ")
  for l in range(0,i+1):
    print("*",end=" ")
  for p in range(i,n):
    print(" ",end=" ")
  print()
```
