```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

```

```python
# lineplot
x = np.arange(0,10,0.1)
y = 2*x +4
print(x,'\n',y)
plt.plot(x,y)
```

```python
x = np.arange(0,10,1)
y = 2*x +4

fig = plt.figure(figsize = (5,5))
plt.plot(x,y, linewidth = 1, linestyle = '--', color = 'b', alpha = 0.9, marker= '*')
plt.title('Line Plot Demo')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Housing Price'], loc = 'upper left')
plt.grid(True)
```

```python
data = {'apples':20, 'mangoes':30, 'lemon':25}

names = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (5,5))
plt.bar(names, values, color = 'orange')

plt.title('Bar Plot Demo')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Bar'], loc = 'upper right')
```

```python
data = {'apples':20, 'mangoes':30, 'lemon':25}

names = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (5,5))
plt.barh(names, values, color = 'orange')

plt.title('Bar Plot Demo')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Bar'], loc = 'lower right')
```

```python
x = np.arange(0,10,1)
y = 2*x +4
y3 = 3*x+7

plt.scatter(x,y, color = 'b', alpha = 0.9, marker= 's', s= 5)
plt.scatter(x,y3,  marker= 'd')
plt.title('Scatter Plot Demo')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Scatter Plot Square', 'Scatter Plot Diamond'], loc = 'upper left')
plt.grid(True)
```

```python
numbers = [10,12,14,15,120,21,30,34,52,15,10,11,22,23,70,80,90,30]

plt.hist(numbers, bins =4, edgecolor = 'black', color = 'r')
plt.title('Histogram Demo')
plt.xlabel('Numbers')
plt.ylabel('Count')
plt.show()
```

```python
total = [10,12,14,15,120, 160,21,30,34,52,15,10,11,22,23,70,80,90,30]
order = [10,12,14,15,120,21,31,33,52,15,115,11,23,25,70,80,99,32]
discount = [11,13,110,15,120,21,33,34,52,15,10,11,22,23,70,80,90,30]

data = list([total,order,discount])

plt.boxplot(data) #used for outliers
plt.show()
```

```python
plt.violinplot(data, showmedians = True) #to see the distribution of your data spread
plt.show()
```

```python
# areaplot
x = np.arange(0,10,0.1)
y = 2*x +4
# plt.plot(x,y, color = 'black')
plt.stackplot(x,y, color = 'black')
```

```python
data = {'dogs':20, 'cats':30, 'lion':25, 'rats':20}

names = list(data.keys())
values = list(data.values())

plt.pie(values, labels = names, autopct='%0.2f%%', explode = [0,0,0,0.2])
plt.show()
```

```python
data = {'dogs':20, 'cats':30, 'lion':25, 'rats':20}

names = list(data.keys())
values = list(data.values())
y = [100000]
plt.pie(values, labels = names, autopct='%0.2f%%')
plt.pie(y,radius = 0.4,colors = 'w')
plt.show()
```

```python
x = np.arange(0,10,1)
y = 2*x +4
y3 = 3*x+7


plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.scatter(x,y3)

plt.subplot(2,2,3)
plt.scatter(y,y3)

plt.subplot(2,2,4)
plt.stackplot(x,y)
```
