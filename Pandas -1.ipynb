```python
d1l = {'Alpha_left':['a','b','c','d'],
       'Number':[1,2,6,5],
       'Alphanumeric_left':['a','b','c','d']
       }
df_l = pd.DataFrame(d1l)
df_l
```

```python
d1r = {'Alpha_right':['a','b','c','d'],
       'Number':[5,6,7,8],
       'Alphanumeric_right':['e5','f6','g7','h8']
       }
df_r = pd.DataFrame(d1r)
df_r
```

```python
df_inner = pd.merge(df_l, df_r, on='Number', how = 'inner')
df_inner
```

```python
df_left = pd.merge(df_l, df_r, on='Number', how = 'left')
df_left
```

```python
df_right = pd.merge(df_l, df_r, on='Number', how = 'right')
df_right
```

```python
df_outer = pd.merge(df_l, df_r, on='Number', how = 'outer')
df_outer
```
