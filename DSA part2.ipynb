#STACKS
**LIFO** : **LAST IN,FIRST OUT**

**Example:** Stack of plates in your sink(Please wash!!!)

FUNCTIONALITY:
- PUSH: adds an item to the top of the stack
- POP: removes the top item from the stack
- PEEK/TOP: see what's at the top without removing it

USES:
- UNDO/REDO Operations
- Backtracking


image.png

```python
class Stack:
  def __init__(self,capacity): #capacity: defines how many elements you can have
    self.stack=[] #using a list to store stack elements
    self.capacity=capacity #None for no limit

  def is_empty(self): #check if stack is empty
    return len(self.stack)==0

  def is_full(self): #check if stack is full
    return len(self.stack)==self.capacity

  def push(self,item): #adding element to the top of the stack
    if self.is_full(): #if stack is full
      print("Stack Overflowing! Can't add more items")
    else:
      self.stack.append(item)
      print(f"Pushed: {item}")

  def pop(self): #removing and returning the popped value
    if self.is_empty():
      print("Stack Underflowing! Nothing to pop")
      return None
    popped_item=self.stack.pop()
    print(f"Popped: {popped_item}")
    return popped_item

  def peek(self): #returns the item at the top of the stack
    if self.is_empty():
      print("Stack Underflowing! Nothing to show")
      return None
    return self.stack[-1]

  def display(self):
    if self.is_empty():
      print("Stack Underflowing! Nothing to show")
      return None
    else:
      print("Stack(top to bottom)")
      for item in reversed(self.stack):
        print(item)

s=Stack(5) #capacity = 5

s.push("Sagar")
s.push("Teena")
s.push("Varsha")
s.push("Daredevil")
s.push("Prakhar")
s.push("Spiderman")

s.display()
s.peek()
s.pop()
s.pop()
s.pop()
s.pop()
s.pop()
s.pop()

s.is_empty()
s.is_full()
```

#Queue
**FIFO - First in, First Out**
**Example:-** An interview (first person to go in is the first person to leave)

FUNCTIONALITY:
- Enqueue : Add an element from the REAR end
- Dequeue : remove an element from the FRONT end


image.png

```python
class Queue:
  def __init__(self,capacity):
    self.queue=[]
    self.capacity=capacity

  def is_empty(self):
    return len(self.queue)==0

  def is_full(self):
    return len(self.queue)==self.capacity

  def Enqueue(self,item): #insertion
   if self.is_full():
    print("Queue Overflowing! Cant add more values")
   else:
    self.queue.append(item) #adding from the REAR END
    print(f"Enqueued: {item}")

  def Dequeue(self):
    if self.is_empty():
      print("Queue Underflowing! Nothing to remove")
    else:
      removed=self.queue.pop(0) #removing from the FRONT END
      print(f"Dequeued: {removed}")

  def display(self):
    if self.is_empty():
      print("Queue Underflowing! Nothing to show")
    else:
      print("Queue(front to rear)")
      for item in self.queue:
        print(item)

q=Queue(3)

q.Enqueue("IRONMAN")
q.Enqueue("Hardik")
q.Enqueue("Pratham")
q.Enqueue("kakashi")
q.display()
q.Dequeue()
q.Dequeue()
q.Dequeue()
q.Dequeue()
```

#Recursion
- A function that calls itself to solve smaller versions of a problem.
- You keep breaking the problem until you reach a base case(end condition), then solve backwards

```python
#Countdown Timer

def Countdown(n): #n=start value of countdown
  if n == 0: #base case
    print("TIME's UP! BOOOM")
  else:
    print(n)
    Countdown(n-1)

Countdown(5)
```

```python
#Factorial - 5! = 5X4X3X2X1

def fact(n):
  if n == 0: #base case
    return 1
  elif n == 1:
    return 1
  return n*fact(n-1)

fact(5)
```

#Search Algorithms

#Linear Search
- Goes through each item one by one until you find the **target**
- **Principle**: One by One
- Slow for long/large lists
- **Example**:
   - BATMAN is to find a pack of chocolates from 10 people standing in a line, so he enquires everyone one by one.

image.png

```python
def linear_search(ppl,target):
  for i in range(len(ppl)):
    if ppl[i] == target:
      print("CULPRIT FOUND! CHOCOLATESSS FOUND")
      return i
  print("Everyone's clean")
  return -1 #not found

people=['Ketan','Prakhar','Nandini','Keerthika','Samikshit']

print(linear_search(people,'Samikshit'))
```

#Binary Search
- Search a **sorted list** by cutting it in half each time
- Much faster than linear search
- **Principle**: Divide and conquer
- **Time Complexity**: O(log n)

image.png

```python
def binary_search(arr,target):
  low=0
  high=len(arr)-1

  while low<=high: #as long as low <= high
    mid=(low+high)//2 #add l+h and divide by 2 to get the middle

    if arr[mid]==target:
      print("TARGET FOUND")
      return mid
    elif arr[mid]<target:
      low=mid+1 #ignore left half including the mid
    else:
      high=mid-1 #ignore right half including the mid

  print("Value not found")
  return -1

arr=[2,4,5,6,12,23,45,50,67,78,88,92]
target=101

index=binary_search(arr,target)

print(f"Target: {target} found at index: {index}")
```

#SORTING ALGORTIHMS

#Bubble Sort
- repeatedly compares adjacent(side by side) elements and swaps them if they are in the wrong order

image.png

```python
#ASCENDING ORDER
def bubble_sort(arr):
  n=len(arr) #getting the number of elements

  for i in range(n): #control how many times the loop will run to compare(n times)
    for j in range(0,n-i-1): #controls how many swaps to do
       if arr[j]>arr[j+1]: #comparing adjacent elements
        arr[j],arr[j+1]=arr[j+1],arr[j] #swap
  return arr

data=[45,10,2,80,52,99,103,24]

sorted_data=bubble_sort(data)
print(f"Sorted Data(Bubble Sort): {sorted_data}")
```

```python
#DESCENDING ORDER
def bubble_sort(arr):
  n=len(arr) #getting the number of elements

  for i in range(n): #control how many times the loop will run to compare(n times)
    for j in range(0,n-i-1): #controls how many swaps to do
       if arr[j]<arr[j+1]: #comparing adjacent elements
        arr[j],arr[j+1]=arr[j+1],arr[j]
  return arr

data=[45,10,2,80,52,99,103,24]

sorted_data=bubble_sort(data)
print(f"Sorted Data(Bubble Sort): {sorted_data}")
```

#Selection Sort
- selects the minimum element from the unsorted part and put it at the beginning.

image.png

```python
def selection_sort(arr):
  n=len(arr)

  for i in range(n):
    minVal=i #ASSUMING i is the minimum value

    for j in range(i+1,n):
      if arr[j]<arr[minVal]: #comparing assumed min with every other value
        minVal=j #if a value smaller than min is found, update min

    arr[i],arr[minVal]=arr[minVal],arr[i] #swapping

  return arr

arr=[15,7,12,2,0,22,50]

sorted_data=selection_sort(arr)
print(f"Sorted data(SELECTION SORT): {sorted_data}")
```

#Insertion Sort:
- simple algorithm that builds the final sorted array one item at a time by comparing the current element with previous ones and inserting it into its correct position.

Insertion-sort-example.png

image.png

```python
def insertion_sort(arr):
  for i in range(1,len(arr)): #starting a loop from i=1
    key=arr[i] #current element, to be inserted at the correct position
    j=i-1 #holds the value before arr[i]

    while j>=0 and key<arr[j]: #if true, means key should come before arr[j]
      arr[j+1]=arr[j] #shifting arr[j] one position to the right to make space for key(current element)
      j-=1 #move to the left by one step

    arr[j+1]=key #placing key in the correct position
  return arr

arr=[5,6,8,2,1]
sorted_data=insertion_sort(arr)
print(f"Sorted data(SELECTION SORT): {sorted_data}")
```

#Merge Sort
- **Principle**: Divide and Conquer
- **Divide** the array into halves
- **Conquer**(sort) each half
- **Merge** the sorted halves

Merge Sort.png

```python
def merge_sort(arr):
  if len(arr)>1:
    mid=len(arr)//2 #finding mid
    left_half=arr[:mid]
    right_half=arr[mid:]

    merge_sort(left_half)
    merge_sort(right_half)

    i=j=k=0 #i: left half, j: right half, k: arr for placing elements back together

    #Merge the 2 halves
    while i<len(left_half) and j<len(right_half):
      if left_half[i]<right_half[j]:
        arr[k]=left_half[i]
        i+=1
      else:
        arr[k]=right_half[j]
        j+=1
      k+=1

    #checking for remaining elements
    while i<len(left_half):
      arr[k]=left_half[i]
      i+=1
      k+=1

    while j<len(right_half):
      arr[k]=right_half[j]
      j+=1
      k+=1

arr=[34,77,80,9,23,50]
merge_sort(arr)
print("Sorted Array(MERGE SORT)",arr)
```

#Quick Sort
- Pick a **pivot**
- **Partition** the array so that:
   -  element < Pivot are on the left side
   -  element > Pivot are on the right side
- Recursion

Quick_sort_algorithm.png

```python
def quick_sort(arr):
  if len(arr)<=1:
    return arr
  else:
    pivot=arr[0] #selecting pivot
    left=[x for x in arr[1:] if x<pivot]
    right=[x for x in arr[1:] if x>pivot]
    return quick_sort(left)+[pivot]+quick_sort(right)

arr=[38,14,60,25,88,90,1,234]
sorted_arr=quick_sort(arr)
print("Sorted Array:", sorted_arr)
```
