```python
# stack - joining mulitple arrays along a new axis
```

```python
a = np.array([10,20,30])
b = np.array([40,50,60])

result = np.stack((a,b)) # axis = 0
print(result)
```

```python
result = np.stack((a,b), axis = 1)
print(result)
```

```python
# hstack ,vstack
```

```python
a = np.array([10,20,30])
b = np.array([40,50,60])
```

```python
result = np.hstack((a,b)) # hstack axis joined along - 1
print(result)
```

```python
result = np.vstack((a,b))  # vstack axis joined along - 0
print(result)
```
