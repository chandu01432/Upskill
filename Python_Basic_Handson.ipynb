4. Write a code to reverse a string.

name='Mansoor'

Output:roosnaM

```python
name='Mansoor'
rev=name[::-1]
print(rev)
```

```python
name='Mansoor'
rev=''
for i in name:
  rev=i+rev
print(rev)
```

```python
name='Mansoor'
i=len(name)-1     #i=7-1,i=6    ('r')
rev=''
while i>=0:
  rev+=name[i]        #name[6]='r'    name[5]='ro'   name[4]='roo'
  i-=1

print(rev)
```

6. Write a code to check whther a number is an armstrong number or not.

ex-153
   1^3+5^3+3^3=153

```python
num=int(input("Enter the number"))
sum=0
temp=num
power=len(str(num))
while temp>0:
  digit=temp%10
  sum+=digit**power
  temp//=10
if sum==num:
  print("It is an armstrong number")
else:
  print("It is not an armstrong number")
```

7.Create a calculator program for addition,subtraction,multiplication and division..Take the input from the user and choice of operation.

```python
def addition(a,b):
  return a+b
def subtraction(a,b):
  return a-b
def multiplication(a,b):
  return a*b
def division(a,b):
  return a/b


a=int(input("Enter the first number"))
b=int(input("Enter the second number"))
choice=input("Enter your choice")


if choice=="+":
  print("Result=",addition(a,b))
elif choice=="-":
  print("Result=",subtraction(a,b))
elif choice=="*":
  print("Result=",multiplication(a,b))
elif choice=="/":
  print("Result=",division(a,b))
else:
  print("Invalid Choice")
```
